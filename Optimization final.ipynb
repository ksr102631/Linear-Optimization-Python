STEP 1 - DEFINING THE PROBLEM
Here we want to maximize the number of talks we want to watch


```python
prob = LpProblem('WatchingTEDTalks',pulp.LpMaximize)
```

```python
data = ted
selected_cols = ['name','event','duration','views']
data = data[selected_cols]
data.reset_index(inplace=True)
```

```python
data.head(3)
```

```python
for a,b in data.iterrows():
    print(a,b)
```

STEP 2 : CREATE THE DECISION VARIABLES

```python
decision_variables = []
for rownum,row in data.iterrows():
    variable = str('x' + str(rownum))
    variable = pulp.LpVariable(variable,lowBound=0,upBound=1,cat="Integer")
    decision_variables.append(variable)
```

```python
print('Total number of decsision variables :'+str(len(decision_variables)))
```

```python
for a,b in enumerate(data):
    print(a,b)
```

```python
for a,b in enumerate(data.values):
    print(a,b)
```

STEP 3 : DEFINE THE OBJECTIVE FUNCTION

```python
for a, b in enumerate(decision_variables):
    print(a,b)
```

WE want to watch only popular movies subject (maximum views) to the constraints the no of days we have, no of movies we want 
to watch

Defining our objective function - maxmizing the popular movies to watch

```python
for rownum,row in data.iterrows():
    print(rownum,row)
```

```python
total_views = ""
for rownum,row in data.iterrows():
    for a,b in enumerate(decision_variables):
        if rownum == a:
            formula = row['views']*b
            total_views += formula
        
```

```python
STEP 4 : Set the constraints
```

```python
total_time_available_for_talks = 10* 60 # 10 hrs to minutes
total_talks_can_watch = 25 # Not to watch too many movies
```

```python
#Constraint 1 - Time for talks
total_time_talks = []
for rownum,row in data.iterrows():
    for a,b in enumerate(decision_variables):
        if rownum == a:
            formula = row['duration']*b
            total_time_talks += formula
```

```python
total_time_talks
```

```python
total_time_talks==total_time_available_for_talks
```

```python
prob += (total_time_talks==total_time_available_for_talks)
```

```python
prob
```

Constraint 2 - Number of Talks


```python
total_talks = ''
for rownum,row in data.iterrows():
    for a,b in enumerate(decision_variables):
        if rownum==a:
            formula =b
            total_talks += formula
            
```

```python
total_talks
```

```python
total_talks == total_talks_can_watch
```

```python
prob += (total_talks == total_talks_can_watch)
```
